```python
#!/usr/bin/python

assault = ['battery', 'assault','threaten','abuse','harm','violent','threat','murder','injury','death','inflict']
drug = ['alc','drunk','marijuana','alcohol','substance','paraphernalia','narcotic','dui','driving','control','controlled','addict','influence','cocaine','liquor']
horrible_other_crimes = ['container','ammunition','manufacture','laser','weapon','concealed','peeping','firearm']
property_crimes = ['theft','vehicle','burglary','vandalism','robbery','rob','vandalize','identification','identity','identific','credit','carjacking','defrauding','defraud','fraudulent','arson','rob','robbery']
sex_crimes = ['body','touch','sexual','rape','copulation','prostitution','intercourse','sodomy','molest','oral','lewd','sex','penetration']

#words = 'MANUFACTURE/SALE/POSSESS/ETC LEADED CANE/BILLY/ETC (F)'
#words = words.lower()
table = str.maketrans("/,:",3*" ")
#words = words.translate(table)
#words = words.split()
#crimes = [assault,drug,property_crimes,sex_crimes]
#words

def standardize_crime(string):
    s = string.lower()
    s = s.translate(table)
    #s = s.strip()
    
    for i in assault:
        if i in s:
            return "ASSAULT"
    for i in drug:
        if i in s:
            return "DRUG"
    for i in horrible_other_crimes:
        if i in s:
            return "HORRIBLE OTHER CRIME"
    for i in property_crimes:
        if i in s:
            return "PROPERTY CRIME"
    for i in sex_crimes:
        if i in s:
            return "SEX CRIME"



df['charge'] = df['charge'].apply(standardize_crime)
#df

            
df.dropna(axis=0, subset=['charge'], inplace=True)

df['charge'].value_counts()

    

    
  
    
    
```
